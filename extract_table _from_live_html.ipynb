## import libraries

```python
import requests
from bs4 import BeautifulSoup
```

## Download html page content

```python
live_url= "https://www.fire.tc.faa.gov/ConeCalorimeterDataMining/listresults.asp?searchList=all&listSubmit=Submit"
response = requests.get(live_url)
response.raise_for_status()
content= response.text  
```

## Apply BeautifulSoup to parse html page

```python
soup = BeautifulSoup(content, 'html.parser')            
all_rows= soup.table.find_all('tr')
```

## Collecting each row from html table

```python
result = []
for row in all_rows[1:]:   
    all_tds= row.findAll('td')
    row= {}
    row['cas_number']= all_tds[1].get_text().replace(" ", "")
    row['material_name']= all_tds[2].get_text().replace(" ", "")
    row['chemical_name']= all_tds[3].get_text().replace(" ", "")
    row['material_type']= all_tds[4].get_text().replace(" ", "")
    row['iso']= all_tds[5].get_text().replace(" ", "")
    row['file_name']= all_tds[6].get_text().replace(" ", "")
    result.append(row)
```

```python
import pandas as pd
pd.DataFrame(result)
```
